```python
# Insert library
import numpy as np 
import matplotlib.pyplot as plt

# Import data
traffic = np.genfromtxt("drive/MyDrive/pengenalan pola/web_traffic.tsv",delimiter='\t')
print(traffic[:10])
print(traffic.shape)

# Copy data in each column
x = traffic[:,0]
y = traffic[:,1]

# Remove NaN data
x = x[~np.isnan(y)]
y = y[~np.isnan(y)]

# Make a plot
plt.scatter(x,y)
plt.title("Web traffic last month")
plt.xlabel("Time")
plt.ylabel("Hits/hour")

plt.xticks([w*7*24 for w in range(10)],['week %i' %w for w in range(10)])
plt.autoscale(tight=True)
plt.grid()
```

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

#Baca gambar
image1 = cv2.imread('drive/MyDrive/pengenalan pola/5.png')
image2 = cv2.imread('drive/MyDrive/pengenalan pola/apel.jpg')
image3 = cv2.imread('drive/MyDrive/pengenalan pola/babon.jpg')
image4 = cv2.imread('drive/MyDrive/pengenalan pola/pokat.jpg')

#Dimensi gambar
height = [image1.shape[0], image2.shape[0], image3.shape[0], image4.shape[0]]
width = [image1.shape[1], image2.shape[1], image3.shape[1], image4.shape[1]]
channel = [image1.shape[2], image2.shape[2], image3.shape[2], image4.shape[2]]
size = [image1.size, image2.size, image3.size, image4.size]

#print
for i in range(len(height)) : 
  print("Dimensi gambar ke-%d : " % (i+1))
  print("Height: %d" % (height[i]))
  print("Width: %d" % (width[i]))
  print("Channel: %d" % (channel[i]))
  print("Size: %d" % (size[i]))
  print()
```

```python
#Ubah ke Grayscale
img_gray = cv2.cvtColor(image4, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
cv2.waitKey(0)
```

```python
#Ubah ke HSV
img_hsv = cv2.cvtColor(image4, cv2.COLOR_BGR2HSV)
cv2_imshow(img_hsv)
cv2.waitKey(0)
```

```python
#Membuat histogram dari citra grayscale
from matplotlib import pyplot as plt
  
#Baca citra dalam grayscale
image1 = cv2.imread('drive/MyDrive/pengenalan pola/5.png', 0)
image2 = cv2.imread('drive/MyDrive/pengenalan pola/apel.jpg', 0)
image3 = cv2.imread('drive/MyDrive/pengenalan pola/babon.jpg', 0)
image4 = cv2.imread('drive/MyDrive/pengenalan pola/pokat.jpg', 0)
  
#Cari frekuensi piksel 
histr = cv2.calcHist([image4],[0],None,[256],[0,256])
  
#tampilkan
plt.plot(histr)
plt.show()
```
